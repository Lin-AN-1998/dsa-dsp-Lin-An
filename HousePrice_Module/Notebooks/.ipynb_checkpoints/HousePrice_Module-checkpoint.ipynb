```python
import app
```

# Read Train Data

```python
input_path = '../data/train.csv'
df_master = app.read_data(input_path)
df_master
```

# Pre-Processing
## Encode categorical variables

```python
one_hot_encoder = app.get_one_hot_encoder(df_master)
encoded_df = app.encode_categorical_data(df_master, one_hot_encoder)
encoded_df
```

## Fill missing value

```python
df_total = app.pre_process(encoded_df)
df_total.isna().sum()
```

# Model Building
## Split data

```python
target_column = 'SalePrice'
X_train, X_test, y_train, y_test = app.split_data(df_total, target_column)
```

## Train model and Make Performance

```python
model = app.train_model(X_train, y_train)
y_pred = model.predict(X_test)
app.compute_rmsle(y_test, y_pred)
```
